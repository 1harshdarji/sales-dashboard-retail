```python
import os
import pandas as pd

file_path = r"C:\Users\darji\Projects\sales-dashboard-retail\data\raw\Sample - Superstore.csv"

if not os.path.exists(file_path):
    print("‚ùå File not found at:", file_path)
else:
    print("‚úÖ File found:", file_path)
    try:
        # Try reading with Windows encoding
        df = pd.read_csv(file_path, encoding='ISO-8859-1')
        print("‚úÖ File loaded successfully!")
        print("Rows:", df.shape[0], "Columns:", df.shape[1])
        display(df.head())
    except UnicodeDecodeError:
        print("üö´ Encoding error: try encoding='latin1' or 'utf-16'.")
    except Exception as e:
        print("‚ö†Ô∏è Unexpected error:", e)
```

```python
# Cell 3: Data Overview
print("Dataset Overview")
print("-" * 60)
print("Shape (rows, columns):", df.shape)
print("\nColumn names:")
print(df.columns.tolist())
print("\nData types:")
print(df.dtypes)
print("\nFirst few rows:")
display(df.head())

```

```python
# Cell 4: Check for missing values
print("Missing Values per Column:")
missing = df.isnull().sum()
print(missing[missing > 0] if missing.sum() > 0 else "‚úÖ No missing values found.")
```

```python
# Cell 5: Check for duplicates
dup_count = df.duplicated().sum()
print(f"Duplicate Rows: {dup_count}")

if dup_count > 0:
    df = df.drop_duplicates()
    print("‚úÖ Duplicates removed. New shape:", df.shape)
else:
    print("‚úÖ No duplicate rows found.")
```

```python
# Cell 6: Convert date columns
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

print("‚úÖ Dates converted to datetime format.")
print(df[['Order Date', 'Ship Date']].dtypes)
```

```python
# Cell 7: Add new time-based features
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Order Month Name'] = df['Order Date'].dt.strftime('%b')

print("‚úÖ Added columns: 'Order Year', 'Order Month', 'Order Month Name'")
display(df[['Order Date', 'Order Year', 'Order Month', 'Order Month Name']].head())
```

```python
# Cell 8: Check unique values in key columns
for col in ['Segment', 'Region', 'Category', 'Sub-Category', 'Ship Mode']:
    print(f"\n{col} - unique values:")
    print(df[col].unique())
```

```python
# Cell 9: Save cleaned dataset
cleaned_path = r"C:\Users\darji\Projects\sales-dashboard-retail\data\cleaned\superstore_cleaned.csv"

# Create folder if it doesn't exist
os.makedirs(os.path.dirname(cleaned_path), exist_ok=True)

df.to_csv(cleaned_path, index=False)
print(f"üíæ Cleaned dataset saved at: {cleaned_path}")
```
